```python
#(train_X, train_y), (test_X, test_y) = mnist.load_data()
Temp_Data = pd.read_csv('C:/Users/pierr/OneDrive/Desktop/Temp_Data.csv')
```

```python
Temp_Data.shape
```

```python

train = int(Temp_Data.shape[0]*.8)
test = int(Temp_Data.shape[0]*.2)

train_X = Temp_Data.iloc[0:train,0]
train_y = Temp_Data.iloc[0:train,1]
test_X = Temp_Data.iloc[0:test,0]
test_y = Temp_Data.iloc[0:test,1]
```

```python
print('X_train: ' + str(train_X.shape))
print('Y_train: ' + str(train_y.shape))
print('X_test:  '  + str(test_X.shape))
print('Y_test:  '  + str(test_y.shape))
```

```python
digit_model = Sequential(
    [
        Dense(1),
        Dense(1,activation ='linear')
        
    ]
)


opt = keras.optimizers.RMSprop(learning_rate = 0.001)

digit_model.compile(loss='mse',
                    optimizer = opt, 
                    metrics = ['accuracy']
                   )


#digit_model.compile(loss = tf.keras.losses.SparseCategoricalCrossentropy() )
#keras.optimizers.RMSprop(learning_rate = 0.001, rho = 0.9)
#digit_model.summary()
```

```python
digit_model.fit(train_X,train_y, epochs = 50)
```

```python
model_output = digit_model.predict(test_X)
```

```python
error = np.zeros(test)
```

```python
for count, x in enumerate(model_output):
    error[count] = model_output[count] - test_y[count]
    if error[count]>0.001:
        print(f"Predicted Value: {model_output[count]} vs Actual Value: {test_y[count]}")
    
```
