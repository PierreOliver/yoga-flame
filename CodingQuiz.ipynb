```python
def solution1(A, B):
    # write your code in Python 3.6
    loopcount = 0
    prod =0
    testrange = range(A,B+1)
    yescount = 0
    
    while prod<B+1:
        prod = loopcount * (loopcount+1)
        loopcount+=1
        if prod in testrange:
            yescount +=1
            
    return yescount    
    
    pass
```

```python
def solution2(A, B):
    # write your code in Python 3.6
    products = []
    prev_x = 0
    loopcount = 0
    prod =0
    
    while prod<B+1:
        prod = loopcount * (loopcount+1)
        products.append(prod)
        loopcount+=1
     
    j = set(range(A,B+1)) & set(products)
    
    if len(j) > 0:
        return len(j)
    else:
        return 0
        
    pass
```

```python
def solution3(A):
   # write your code in Python 3.6
    diff = []
    temp = 1
    
    if len(A)<=1:
        return False
    
    A.sort()
    
    for x in A:
        value = A[temp]-x
        if value ==1:
            diff.append(A[temp]-x)
            
        temp+=1
        
        if temp > (len(A)-1):
            break
      
    if len(diff) >= 1:
        return True
    else:
        return False
    pass
```

```python
def solution4(N):
    # write your code in Python 3.6
    solution =[]
    
    if N%2 ==0:
        for x in range(1,int(N/2)+1):
            solution.append(x)
            solution.append(-x)
    else:
        for x in range(1,int(N/2)+1):
            solution.append(x)
            solution.append(-x)
        solution.append(0)
    
    return solution
    pass
```
